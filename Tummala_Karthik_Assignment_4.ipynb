```python


newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
data = random.sample(newsgroups.data, 2000)


stop_words = set(stopwords.words('english'))

def preprocess(doc):
    tokens = word_tokenize(doc.lower())
    return [token for token in tokens if token.isalpha() and token not in stop_words]

processed_docs = [preprocess(doc) for doc in data]

dictionary = Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

tfidf_vectorizer = TfidfVectorizer(tokenizer=lambda doc: doc, lowercase=False)
count_vectorizer = CountVectorizer(tokenizer=lambda doc: doc, lowercase=False)

tfidf = tfidf_vectorizer.fit_transform(processed_docs)
count = count_vectorizer.fit_transform(processed_docs)

def compute_coherence_values(model_type='lda', texts=processed_docs, dictionary=dictionary,
                              corpus=corpus, vectorizer_data=None, start=2, limit=15, step=1):
    coherence_scores = []
    models = []
    for k in range(start, limit + 1, step):
        if model_type == 'lda':
            model = LatentDirichletAllocation(n_components=k, random_state=42)
            model.fit(vectorizer_data)
            topics = [[word for word, _ in sorted(zip(count_vectorizer.get_feature_names_out(), topic), key=lambda x: -x[1])[:10]]
                      for topic in model.components_]
        elif model_type == 'lsa':
            model = TruncatedSVD(n_components=k, random_state=42)
            model.fit(vectorizer_data)
            topics = [[word for word, _ in sorted(zip(tfidf_vectorizer.get_feature_names_out(), comp), key=lambda x: -x[1])[:10]]
                      for comp in model.components_]
        cm = CoherenceModel(topics=topics, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_scores.append(cm.get_coherence())
        models.append(model)
    return models, coherence_scores

lda_models, lda_scores = compute_coherence_values('lda', vectorizer_data=count)
lsa_models, lsa_scores = compute_coherence_values('lsa', vectorizer_data=tfidf)

x = list(range(2, 16))
plt.figure(figsize=(10, 5))
plt.plot(x, lda_scores, label='LDA Coherence', marker='o')
plt.plot(x, lsa_scores, label='LSA Coherence', marker='x')
plt.xlabel("Number of Topics (k)")
plt.ylabel("Coherence Score")
plt.title("LDA vs LSA Topic Coherence Scores")
plt.legend()
plt.grid()
plt.show()

best_k_lda = x[np.argmax(lda_scores)]
best_k_lsa = x[np.argmax(lsa_scores)]
print(f"Optimal number of topics (LDA): {best_k_lda}, Coherence: {max(lda_scores):.4f}")
print(f"Optimal number of topics (LSA): {best_k_lsa}, Coherence: {max(lsa_scores):.4f}")

def print_topics(model, vectorizer, k):
    terms = vectorizer.get_feature_names_out()
    for i, comp in enumerate(model.components_):
        top_words = [terms[j] for j in comp.argsort()[-10:]]
        print(f"Topic {i+1}: {' | '.join(top_words[::-1])}")

print("\n--- LDA Topics ---")
print_topics(lda_models[np.argmax(lda_scores)], count_vectorizer, best_k_lda)

print("\n--- LSA Topics ---")
print_topics(lsa_models[np.argmax(lsa_scores)], tfidf_vectorizer, best_k_lsa)
```
